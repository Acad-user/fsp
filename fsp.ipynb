```python
#recursive factorial calculation function
def factorial(n):
  if(n==0 or n==1): return 1    #base case : for some n values outputs are pre-known to us
  return factorial(n-1) * n     #recursive case:the function will call itself either directly or indirectly
print(factorial(2))
print(factorial(5))
print(factorial(6))
```

```python
# Example 1:
# Input: words = ["Hello","Alaska","Dad","Peace"]
# Output: ["Alaska","Dad"]
# Explanation:
# Both "a" and "A" are in the 2nd row of the American keyboard due to case insensitivity.

# Example 2:
# Input: words = ["omk"]
# Output: []

# Example 3:
# Input: words = ["adsdf","sfd"]
# Output: ["adsdf","sfd"]

def findWords(words):

    row1 = set("qwertyuiop")
    row2 = set("asdfghjkl")
    row3 = set("zxcvbnm")


    row_dict = {
        **{char: 1 for char in row1},
        **{char: 2 for char in row2},
        **{char: 3 for char in row3}
    }

    result = []

    for word in words:

        low_word = word.lower()
        first_row = row_dict[low_word[0]]


        if all(row_dict[char] == first_row for char in low_word):
            result.append(word)

    return result


words1 = ["Hello", "Alaska", "Dad", "Peace"]
print(findWords(words1))

words2 = ["omk"]
print(findWords(words2))

words3 = ["adsdf", "sfd"]
print(findWords(words3))
```

```python
def select_words(mywords):
    row1 = set("qwertyuiop")
    row2 = set("asdfghjkl")
    row3 = set("zxcvbnm")
    selected_words=[]
    for word in mywords:
        low_word = word.lower()
        #Create the wordset here for each word
        wordset = set(low_word)
        if(wordset.issubset(row1) or wordset.issubset(row2) or wordset.issubset(row3)):
            selected_words.append(word)
    return selected_words

mywords = ["Hello", "Alaska", "Dad", "Peace"]
print(select_words(mywords))
print(select_words(["omk"]))
print(select_words(["adsdf", "sfd"]))
```

```python
def is_palindrome(s):
    # Convert the string to lowercase for case-insensitive comparison
    s = s.lower()
    # A string is a palindrome if it reads the same forwards and backwards
    return s == s[::-1]

def find_palindromes(strings):
    # Use a set comprehension to filter palindromes
    return {s for s in strings if is_palindrome(s)}

# Example set of strings
string_set = {"Radar", "hello", "level", "world", "python", "deified"}

# Finding palindromes in the set (case-insensitive)
palindromes = find_palindromes(string_set)

print("Palindromes found:", palindromes)
```

```python
class Stack:
    def __init__(self, max_size):
        self.stack = []
        self.max_size = max_size

    def push(self, data):
        if len(self.stack) < self.max_size:
            self.stack.append(data)
        else:
            print("Stack is full!")

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        else:
            print("Stack is empty!")
            return None

    def display(self):
        if self.is_empty():
            print("Stack is empty!")
        else:
            print("Stack elements:", self.stack)

    def find_average(self):
        if len(self.stack) == 0:
            print("Stack is empty! Cannot compute average.")
            return None
        return sum(self.stack) / len(self.stack)

    def is_empty(self):
        return len(self.stack) == 0


if __name__ == "__main__":
    def menu_display():
        choice = 0
        while choice < 1 or choice > 5:
            print("\nMAIN MENU")
            print("-----------")
            print("1.Push")
            print("2.Pop")
            print("3.Display")
            print("4.Average")
            print("5.Exit")
            try:
                choice = int(input("Enter your choice: "))
            except ValueError:
                print("Invalid input! Please enter a number between 1 and 5.")
        return choice

    max_size = int(input("Enter the size of the stack: "))
    stk = Stack(max_size)

    choice = 0
    while choice != 5:
        choice = menu_display()
        if choice == 1:
            data = int(input("Enter the data item to be pushed: "))
            stk.push(data)
        elif choice == 2:
            popped_item = stk.pop()
            if popped_item is not None:
                print("Popped item is:", popped_item)
        elif choice == 3:
            stk.display()
        elif choice == 4:
            avg = stk.find_average()
            if avg is not None:
                print("Average of all items is:", avg)
        elif choice == 5:
            print("Exiting...")
        else:
            print("Invalid choice!")
```

```python
class Node:
    def __init__(self, info):
        self.__data = info
        self.__next = None

    def get_data(self):
        return self.__data

    def set_data(self, info):
        self.__data = info

    def get_next(self):
        return self.__next

    def set_next(self, next_node):
        self.__next = next_node


class LinkedList:
    def __init__(self):
        self.__head = None
        self.__tail = None

    def get_head(self):
        return self.__head

    def set_head(self, node):
        self.__head = node

    def get_tail(self):
        return self.__tail

    def set_tail(self, node):
        self.__tail = node

    def append(self, info):
        new_node = Node(info)
        if self.get_head() is None:
            self.set_head(new_node)
            self.set_tail(new_node)
            new_node.set_next(None)
            print("Creating the Linked List...")
        else:
            print("Appending the new node at the end...")
            self.get_tail().set_next(new_node)
            self.set_tail(new_node)

    def display(self):
        ptr = self.get_head()
        if ptr is None:
            print("Linked List is not pre-existing...")
        else:
            print("Displaying the content of the Linked List")
            while ptr is not None:
                print(ptr.get_data(), end=", ")
                ptr = ptr.get_next()
            print()

    def __str__(self):
        ptr = self.get_head()
        if ptr is None:
            return "Linked List is not pre-existing..."
        msg = []
        while ptr is not None:
            msg.append(str(ptr.get_data()))
            ptr = ptr.get_next()
        return ' '.join(msg)

    def search(self, info):
        ptr = self.get_head()
        if ptr is None:
            print("Linked List is not pre-existing...")
            return False
        while ptr is not None:
            if ptr.get_data() == info:
                print(f"Found: {info}")
                return True
            ptr = ptr.get_next()
        print(f"Not found: {info}")
        return False


# === Test Code ===
lnk_list = LinkedList()
lnk_list.display()
print(lnk_list)

info_data = [11, 22, 33, 44, 55]
for item in info_data:
    lnk_list.append(item)

lnk_list.display()
print(lnk_list)
lnk_list.search(100)
lnk_list.search(33)
```
